```python
import os
import time
import requests
import base64
import io
from PIL import Image
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Setup Chrome options
chrome_options = Options()
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--headless")
driver = webdriver.Chrome(options=chrome_options)
```

```python
categories = ["cat","dog","car","mountains", "flowers", "airplanes",
    "bicycles", "birds", "boats", "bridges", "butterflies",
    "castles", "clouds", "deserts", "elephants", "fireworks",
    "forests", "fruits", "lakes", "trains"]

metadata = []

for category in categories:
    # Create directory structure
    os.makedirs(f"multicategory_image_dataset/images/{category}", exist_ok=True)
    
    # Search Google Images
    driver.get(f"https://www.google.com/search?q={category}&tbm=isch")
    
    # Accept cookies if prompted (this might be necessary in some regions)
    try:
        WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Accept all')]"))
        ).click()
    except:
        pass  # No cookie prompt or different text
    
    # Implement progressive scrolling to load images
    image_count = 0
    last_height = driver.execute_script("return document.body.scrollHeight")
    
    while image_count < 50:
        # Scroll down gradually to trigger lazy loading
        for scroll_step in range(10):
            driver.execute_script(f"window.scrollTo(0, {scroll_step * last_height / 10});")
            time.sleep(0.5)
        
        # Scroll to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        
        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            # Try clicking "Show more results" if available
            try:
                show_more = driver.find_element(By.CSS_SELECTOR, ".mye4qd")
                if show_more.is_displayed():
                    show_more.click()
                    time.sleep(2)
                else:
                    break
            except:
                break
        last_height = new_height
        
        # Try multiple selectors to find images
        selectors = [
            "img.rg_i", 
            "img.Q4LuWd", 
            "img.YQ4gaf", 
            "img[data-src]",
            "img[src]"
        ]
        
        for selector in selectors:
            img_elements = driver.find_elements(By.CSS_SELECTOR, selector)
            if img_elements:
                break
        
        # Extract image URLs, checking both src and data-src attributes
        img_urls = []
        for img in img_elements:
            # Check for src attribute
            src = img.get_attribute('src')
            if not src:
                # Try data-src if src is not available
                src = img.get_attribute('data-src')
            if not src:
                # Try other attributes like data-original
                src = img.get_attribute('data-original')
            
            if src and src not in img_urls:
                img_urls.append(src)
        
        # Download images (only new ones)
        current_count = image_count
        for img_url in img_urls:
            if image_count >= 50:
                break
                
            try:
                # Handle base64 encoded images
                if img_url.startswith('data:image'):
                    # Extract the base64 string
                    img_data = img_url.split(',')[1]
                    # Decode the base64 string
                    img_bytes = base64.b64decode(img_data)
                    img = Image.open(io.BytesIO(img_bytes))
                    filename = f"multicategory_image_dataset/images/{category}/img_{image_count}.jpg"
                    img.save(filename)
                    resolution = f"{img.width}x{img.height}"
                else:
                    # Regular URL
                    response = requests.get(img_url, timeout=10)
                    if response.status_code == 200:
                        filename = f"multicategory_image_dataset/images/{category}/img_{image_count}.jpg"
                        with open(filename, "wb") as f:
                            f.write(response.content)
                        
                        # Get image resolution
                        img = Image.open(filename)
                        resolution = f"{img.width}x{img.height}"
                    else:
                        continue
                
                # Add to metadata
                metadata.append({
                    "category": category,
                    "url": img_url,
                    "filename": filename,
                    "resolution": resolution
                })
                
                image_count += 1
                print(f"Downloaded image {image_count}/50 for category '{category}'")
                
            except Exception as e:
                print(f"Error downloading {img_url}: {e}")
        
        # If no new images were downloaded in this iteration, break the loop
        if current_count == image_count:
            print(f"No new images found after scrolling. Stopping at {image_count} images.")
            break
```

```python
# Save metadata to CSV
pd.DataFrame(metadata).to_csv(f"multicategory_image_dataset/metadata.csv", index=False)
driver.quit()

print(f"Downloaded {len(metadata)} images. Metadata saved to multicategory_image_dataset/metadata.csv")
```
