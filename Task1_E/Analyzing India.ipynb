```python
import pandas as pd

# Load the dataset
df = pd.read_csv("livestock_census_2019_Andhra_pradesh.csv")

# Display the first few rows
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Calculate summary statistics
summary_stats = df.describe()
print(summary_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot distribution of cattle across districts
plt.figure(figsize=(10, 6))
sns.barplot(x='district_name', y='cattle', data=df)
plt.title('Cattle Distribution Across Districts')
plt.xlabel('District')
plt.ylabel('Number of Cattle')
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='district_name', y='buffalo', data=df)
plt.title('Buffalo Distribution Across Districts')
plt.xlabel('District')
plt.ylabel('Number of buffaloes')
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='district_name', y='sheep', data=df)
plt.title('Sheep Distribution Across Districts')
plt.xlabel('District')
plt.ylabel('Number of Sheep')
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='district_name', y='goat', data=df)
plt.title('Goat Distribution Across Districts')
plt.xlabel('District')
plt.ylabel('Number of goats')
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='district_name', y='Total_poultry', data=df)
plt.title('Poultry Distribution Across Districts')
plt.xlabel('District')
plt.ylabel('Total poultry')
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='district_name', y='pig', data=df)
plt.title('Pigs Distribution Across Districts')
plt.xlabel('District')
plt.ylabel('Number of pigs')
plt.xticks(rotation=90)
plt.tight_layout()
```

```python

# Plot correlation heatmap
corr_matrix = df[['cattle','buffalo', 'sheep', 'goat', 'pig','Total_poultry']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Livestock Counts')
plt.tight_layout()
```
